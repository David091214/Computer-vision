```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python
image = cv2.imread("./trial.JPG")
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(rgb_image) # will look bgr #if you want to 
```

```python
lower_white = np.array([10,0,150])
upper_white = np.array([15,140,231])
```

```python
mask = cv2.inRange(hsv_image, lower_white, upper_white)
```

```python
plt.imshow(mask)
```

```python
black_hsv = np.uint8([[[120, 255, 0]]]) 
black_bgr = cv2.cvtColor(black_hsv, cv2.COLOR_HSV2BGR)[0][0]
```

```python
image_result = image.copy()
```

```python
image_result[mask>0] = black_bgr
```

```python
image_result_rgb = cv2.cvtColor(image_result, cv2.COLOR_BGR2RGB)
```

```python
#plot 

fig, axs = plt.subplots(1,3, figsize=(12,6))

axs[0].imshow(rgb_image)
axs[0].set_title("original image")
axs[0].axis('off')

axs[1].imshow(mask)
axs[1].set_title("Binary Mask ")
axs[1].axis('off')

axs[2].imshow(image_result_rgb)
axs[2].set_title("Binary Mask ")
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
