```python
import cv2
```

```python
video = cv2.VideoCapture("./mountain.mp4") 
# to read video we use cv2.VideoCapture
```

```python
#display video 
while True:
    ret, frame = video.read()
    if not ret:
        break
    cv2.imshow("video Frame", frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows
```

```python
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
```

```python
# so now our VideoWriter format is mp4v
```

```python
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))

out = cv2.VideoWriter("./New.mp4", fourcc,30, (width, height))

while True:
    ret , frame = video.read()
    if not ret:
        break
    frame = frame + 50
    out.write(frame)

out.release()
video.release()
cv2.destroyAllWindows
```
